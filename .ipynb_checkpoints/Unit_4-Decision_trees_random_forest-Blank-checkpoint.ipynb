```python
# gini
R1 = np.sort(R1)
R2 = np.sort(R2)

r1mean = R1.mean()
r2mean = R2.mean()

diffsum1 = np.sum(np.absolute(R1[:, None] - R1))
diffsum2 = np.sum(np.absolute(R2[:, None] - R2))


gini1 = diffsum1 / (2 * len(R1)**2 * r1mean)
gini2 = diffsum2 / (2 * len(R2)**2 * r2mean)


gini1, gini2
```

```python
# gini

def calc_gini(R):
    R = np.sort(R)

    mean = R.mean()
    diffsum = np.sum(np.absolute(R[:, None] - R))

    gini = diffsum / (2 * len(R)**2 * mean)
    return gini
```

```python

def normalize_array(arr):
    norm_arr = (arr - np.min(arr)) / (np.max(arr) - np.min(arr))
    return norm_arr
```

```python
P1 = normalize_array(R1)

entropy1 = -np.sum(P1 * np.log2(P1, where = a != 0))

entropy1
```

```python
def calc_entropy(R):
    P = normalize_array(R)
    entropy = -np.sum(P * np.log2(P, where = P != 0))
    return entropy
```

```python
gini_weight = 0.5
entropy_weight = 0.5

# Calculate weighted sum
weighted_sum1 = gini_weight * calc_gini(R1) + entropy_weight * calc_entropy(R1)


weighted_sum2 = gini_weight * calc_gini(R2) + entropy_weight * calc_entropy(R2)

weighted_sum1, weighted_sum2
```
